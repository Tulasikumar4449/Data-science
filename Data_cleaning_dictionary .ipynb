```python
import numpy as np
import pandas as pd
```

```python
data={'temperature':[1,np.nan,2,3,4,np.nan,5,6,np.nan,7],
     'humidity':[22,np.nan,45,np.nan,33,66,np.nan,22,11,10]}
```

```python
ds=pd.DataFrame(data)
```
